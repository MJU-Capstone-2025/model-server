```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('sliding_window_predictions.csv')
print(df.head())
```

```python
window_3 = df[df['window_id']==3]
print(window_3)
```

```python
# date 컬럼을 datetime 형식으로 변환
window_3['date'] = pd.to_datetime(window_3['date'])

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# prediction과 actual 데이터 플로팅
plt.plot(window_3['date'], window_3['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
plt.plot(window_3['date'], window_3['actual'], label='Actual', marker='x', linestyle='-', color='blue')

# 그래프 제목과 레이블 설정
plt.title('Prediction vs Actual for Window ID 3')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)

# x축 날짜 포맷 설정
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
window_97 = df[df['window_id']==97]
print(window_97)
```

```python
# date 컬럼을 datetime 형식으로 변환
window_97['date'] = pd.to_datetime(window_97['date'])

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# prediction과 actual 데이터 플로팅
plt.plot(window_97['date'], window_97['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
plt.plot(window_97['date'], window_97['actual'], label='Actual', marker='x', linestyle='-', color='blue')

# 그래프 제목과 레이블 설정
plt.title('Prediction vs Actual for Window ID 207')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)

# x축 날짜 포맷 설정
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
# window_3과 window_97만 시각화. subplot을 사용하여 두 개의 그래프를 나란히 표시
fig, axs = plt.subplots(1, 2, figsize=(15, 6))
# window_3 그래프
axs[0].plot(window_3['date'], window_3['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
axs[0].set_title('Prediction vs Actual for Window ID 3')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Price')
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
axs[0].xaxis.set_major_locator(mdates.AutoDateLocator())
axs[0].tick_params(axis='x', rotation=45)
axs[0].legend()
# window_97 그래프
axs[1].plot(window_97['date'], window_97['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
axs[1].set_title('Prediction vs Actual for Window ID 97')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Price')
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
axs[1].xaxis.set_major_locator(mdates.AutoDateLocator())
axs[1].tick_params(axis='x', rotation=45)
axs[1].legend()
# 그래프 레이아웃 조정
plt.tight_layout()
# 그래프 표시
plt.show()
```

```python
# 모든 window를 하나의 그래프에 표시
fig, ax = plt.subplots(figsize=(15, 6))
for window_id in df['window_id'].unique():
    window_data = df[df['window_id'] == window_id]
    ax.plot(window_data['date'], window_data['prediction'], label=f'Prediction (Window {window_id})', marker='o', linestyle='--')
    ax.plot(window_data['date'], window_data['actual'], label=f'Actual (Window {window_id})', marker='x', linestyle='-')
# 그래프 제목과 레이블 설정
ax.set_title('Prediction vs Actual for All Windows')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.grid(True, linestyle='--', alpha=0.7)
# x축 날짜 포맷 설정
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.tick_params(axis='x', rotation=45)
# 범례 표시
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small')
# 그래프 레이아웃 조정
plt.tight_layout()
# 그래프 표시
plt.show()
```
