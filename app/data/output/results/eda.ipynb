```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('sliding_window_predictions.csv')
print(df.head())
```

```python
window_1 = df[df['window_id']==1]
print(window_1)
```

```python
# date 컬럼을 datetime 형식으로 변환
window_1['date'] = pd.to_datetime(window_1['date'])

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# prediction과 actual 데이터 플로팅
plt.plot(window_1['date'], window_1['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
plt.plot(window_1['date'], window_1['actual'], label='Actual', marker='x', linestyle='-', color='blue')

# 그래프 제목과 레이블 설정
plt.title('Prediction vs Actual for Window ID 1')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)

# x축 날짜 포맷 설정
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
window_89 = df[df['window_id']==89]
print(window_89)
```

```python
# date 컬럼을 datetime 형식으로 변환
window_89['date'] = pd.to_datetime(window_89['date'])

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# prediction과 actual 데이터 플로팅
plt.plot(window_89['date'], window_89['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
plt.plot(window_89['date'], window_89['actual'], label='Actual', marker='x', linestyle='-', color='blue')

# 그래프 제목과 레이블 설정
plt.title('Prediction vs Actual for Window ID 89')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)

# x축 날짜 포맷 설정
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=45)

# 범례 표시
plt.legend()

# 그래프 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

```python
# window_3과 window_97만 시각화. subplot을 사용하여 두 개의 그래프를 나란히 표시
fig, axs = plt.subplots(1, 2, figsize=(15, 6))
# window_3 그래프
axs[0].plot(window_1['date'], window_1['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
axs[0].set_title('Prediction vs Actual for Window ID 1')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Price')
axs[0].grid(True, linestyle='--', alpha=0.7)
axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
axs[0].xaxis.set_major_locator(mdates.AutoDateLocator())
axs[0].tick_params(axis='x', rotation=45)
axs[0].legend()
# window_97 그래프
axs[1].plot(window_89['date'], window_89['prediction'], label='Prediction', marker='o', linestyle='--', color='red')
axs[1].set_title('Prediction vs Actual for Window ID 89')
axs[1].set_xlabel('Date')
axs[1].set_ylabel('Price')
axs[1].grid(True, linestyle='--', alpha=0.7)
axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
axs[1].xaxis.set_major_locator(mdates.AutoDateLocator())
axs[1].tick_params(axis='x', rotation=45)
axs[1].legend()
# 그래프 레이아웃 조정
plt.tight_layout()
# 그래프 표시
plt.show()
```

```python
# 모든 window를 하나의 그래프에 표시
fig, ax = plt.subplots(figsize=(15, 6))
for window_id in df['window_id'].unique():
    window_data = df[df['window_id'] == window_id]
    ax.plot(window_data['date'], window_data['prediction'], label=f'Prediction (Window {window_id})', marker='o', linestyle='--')
    ax.plot(window_data['date'], window_data['actual'], label=f'Actual (Window {window_id})', marker='x', linestyle='-')
# 그래프 제목과 레이블 설정
ax.set_title('Prediction vs Actual for All Windows')
ax.set_xlabel('Date')
ax.set_ylabel('Price')
ax.grid(True, linestyle='--', alpha=0.7)
# x축 날짜 포맷 설정
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.tick_params(axis='x', rotation=45)
# 범례 표시
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small')
# 그래프 레이아웃 조정
plt.tight_layout()
# 그래프 표시
plt.show()
```

```python
# 81~90까지의 window_id를 가진 데이터만 필터링
window_81_90 = df[df['window_id'].between(81, 90)]
# 그래프 크기 설정
plt.figure(figsize=(15, 6))
# prediction과 actual 데이터 플로팅
for window_id in window_81_90['window_id'].unique():
    window_data = window_81_90[window_81_90['window_id'] == window_id]
    plt.plot(window_data['date'], window_data['prediction'], label=f'Prediction (Window {window_id})', marker='o', linestyle='--')
    plt.plot(window_data['date'], window_data['actual'], label=f'Actual (Window {window_id})', marker='x', linestyle='-')
# 그래프 제목과 레이블 설정
plt.title('Prediction vs Actual for Windows 81 to 90')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True, linestyle='--', alpha=0.7)
# x축 날짜 포맷 설정
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=45)
# 범례 표시
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='small')
# 그래프 레이아웃 조정
plt.tight_layout()
# 그래프 표시
plt.show()
```
